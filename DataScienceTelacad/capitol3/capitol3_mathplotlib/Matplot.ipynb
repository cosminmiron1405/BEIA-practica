```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.title('Titlu')   #titlu plot
plt.xlabel('Axa Ox')  #titlu axa Ox
plt.ylabel('Axa Oy')  #titlu axa Oy 
plt.axis([0,6,0,20])  # seteaza mai intai capetele axei Ox (de la 0 la 6) dupa seteaza limitele axei Oy (de la 0 la 20)
plt.show()  #afisarea plot ului
```

```python
#exemplu plot

import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4])
y1 = x*2
y2 = x**2   # x*x

plt.figure(figsize = (10,5))
plt.plot(x, y1, color = "black", linestyle='dashdot', linewidth=2, marker='o', markerfacecolor='green', markersize=12)
plt.plot(x, y2, color = "red", linestyle='solid', linewidth=2, marker='h', markerfacecolor='yellow', markersize=12)

plt.legend(['x*2', 'x**2'], loc='lower right')
plt.show()
```

```python
#exemplu subplot
import matplotlib.pyplot as plt

plt.subplot(1,2,1)
plt.plot([1,2,3,4], [2,4,9,16])

plt.subplot(1,2,2)
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])

plt.show()
```

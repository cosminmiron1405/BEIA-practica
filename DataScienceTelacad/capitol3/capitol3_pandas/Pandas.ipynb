CURATAREA DATELOR

```python
#stergerea randurilor/coloanelor
df.drop(['parent_salary','house_area'], axis = 1, inplace = True)
#axis => axis = 1 -> coloane
    #    axis = 0 -> randuri
#inplace = False -> returneaza o copie modificata a DataFramelului
#inplace = True -> returneaza DataFrame ul modificat definitiv, modificarile sunt facute permanent
df
```

```python
#stergerea randurilor ale caror valori sunt duplicate
df.drop_duplicates(["type_school"])
```

```python
#stergerea randurilor/coloanelor ale caror celule au valori lipsa
df.dropna()
```

```python
#inlocuirea valorilor nule pe randuri/coloane
df.fillna('fill_value')
```

```python
#inlocuirea valorilor deja existente cu o valoare noua
df.replace('Less Interested', 'Not Interested')
```

```python
#redenumirea indecsilor randurilor sau numelor coloanelor
df.rename(columns={'type_school':'tipul scolii', 'school_accreditation':'acreditarea_scolii'})
```

GRUPAREA DATELOR

```python
#gruparea datelor din setul de date
df.groupby(['type_school']).agg('count')
#cu funtia pusa in continuare cu .agg('?'), intre acele paranteze se poate pune orice 
#operatie de la count, min, max la mean, etc.
```

IMBINAREA DATELOR

```python
#pentru a folosi 2 df uri creem unul identic cu cel utilizat pana acum
df1 = df
df1
```

```python
#unirea datelor in functe de una sau mai multe coloane comune
pd.merge(df, df1, how='inner', on=['type_school'])
#unirea in functei de una sau mai multe coloane
#
```

```python
#concatenarea obiectelor in Pandas (de cele mai multe ori dataframe-uri) de-a lungul unei axe
pd.concat([df,df1], axis = 1)
#axis = 1 -> da modul de alaturare la coloane
#axis = 0 -> da modul de alaturare la randuri
```

```python
pd.concat([df,df1], axis = 0)
```

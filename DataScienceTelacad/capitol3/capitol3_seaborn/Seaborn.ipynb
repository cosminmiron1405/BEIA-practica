IMPORTAREA BIBLIOTECII

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

TIPURI DE REPREZENTARI
ma voi folosi de datele:
-x = variabila de pe axa Ox;
-y = variabila de pe axa Oy;
-data = intregul set de date pus la dispozitie pentru a experimenta

REPREZANTARI GRAFICE PENTRU DISTRIBUTII

```python
#displot
sns.displot(x = tips['tip'])
```

```python
#jointplot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter') #1
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg') #2
```

```python
#pairplot
sns.pairplot(data=tips)
```

REPREZENTARI PENTRU VARIABILE CATEGORICE


```python
#barplot
sns.barplot(x='sex', y='tip', data = tips)
```

```python
#countplot
sns.countplot(x='sex',data=tips)
```

```python
#boxplot
sns.boxplot(x="day", y="tip", data=tips)
```

```python
#violinplot
sns.violinplot(x="day", y="tip", data=tips)
```

```python
#stripplot
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
#swarmplot
sns.swarmplot(x="day", y="total_bill", data=tips)
```

REPREZENTARI SUB FORMA DE MATRICE

```python
#heatmap
sns.heatmap(tips.corr(x='total_bill',y='tip',data=tips))
```

```python
#lmplot
sns.lmplot(x='total_bill',y='tip',data=tips)
```
